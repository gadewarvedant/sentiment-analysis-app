TEXT PREPROCESSING 

```python
#converting to lowercase

df["review"] = df["review"].str.lower()
```

```python
# removing urls
#for removing urls we use re.sub(it is a regular expression library built to handle text data)

import re

def remove_urls(text):
    text = re.sub(r"http\S" , "", text)
    return text
df["review"]= df["review"].apply(remove_urls)
```

```python
# removing punctuation
def remove_puncuation(text):
    text = re.sub(r"[^A-Za-z0-9\s]","",text)
    return text
df["review"]=df["review"].apply(remove_puncuation)
```

```python
df.head(8)
```

```python
#removing html

def remove_html(text):
    text = re.sub(r"<,*?>","",text)
    return text
df["review"]=df["review"].apply(remove_html)
```

REMOVING STOPWORDS

```python
#for stopword removing we use NLTK library which is use in natural language processing
```

```python
import nltk

nltk.download("punkt")
nltk.download("punkt_tab")
nltk.download("stopwords")
```

```python
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
```

```python
def remove_stopword(text):
    tokens = word_tokenize(text)
    stop_words = stopwords.words("english") #nltk has already has stopwords defined (a,an,the....) 

    for word in tokens:
        if word in stop_words:
            text = text.replace(word, "")

    return text
df["review"] = df["review"].apply(remove_stopword)
```

```python
#stemming-when we apply stemming it returns only prefix word for eg (respectful-respect)

from nltk import PorterStemmer #we apply stemming process by using porterstemer algorithm
```

```python
def apply_stemming(text):
    ps = PorterStemmer()
    stemmed_word = []

    tokens = word_tokenize(text)
    for token in tokens:
        stemmed_tokens = ps.stem(token)
        stemmed_word.append(stemmed_tokens)
    return " ".join(stemmed_word)
df["review"] = df["review"].apply(apply_stemming)
```

```python
import torch
from torch.utils.data import DataLoader,TensorDataset
```

```python
train_set = TensorDataset(
    torch.from_numpy(x_train).float(),
    torch.from_numpy(y_train).float()
)

test_set = TensorDataset(
    torch.from_numpy(x_test).float(),
    torch.from_numpy(y_test).float()
)


```

```python
train_loader = DataLoader(train_set, shuffle=True, batch_size=32)
test_loader = DataLoader(test_set, shuffle=True, batch_size=32)
```

```python
import torch.nn as nn 
import torch.optim as optim
```

```python
class RNN(nn.Module):
    def __init__(self,input_size,hidden_size=128,num_layers=1):
        super().__init__()
        self.hidden_size = hidden_size
        self.num_layers = num_layers

        self.rnn= nn.RNN(input_size,hidden_size,num_layers,batch_first=True) #RNN layer
        self.dropout = nn.Dropout(0.5)
        self.fc = nn.Linear(hidden_size,1) #fully connected layer

    def forward(self, x):
        # optional => shape (num of layers, batch size, hidden size)
        h0 = torch.zeros(self.num_layers, x.size(0), self.hidden_size)

        out, _ = self.rnn(x, h0) 
        # 1st value = hidden state of all the timesteps => (batch, seq_len, hidden size)
        # 2nd value = final hidden state of last timestep
        x = self.dropout(x)
        out = self.fc(out[:, -1, :])
        return out

```

```python
input_size = x_train.shape[1]
model = RNN(input_size)

criterion = nn.BCELoss()
optimizer = torch.optim.Adam(model.parameters(), lr=0.001, weight_decay=1e-4)
```

```python
# epochs = 15

# for epoch in range(epochs):
#     model.train()

#     correct_vals = 0
#     tot_vals = 0
#     total_loss = 0

#     for Xb, yb in train_loader:
#         optimizer.zero_grad()

#         Xb = Xb.unsqueeze(1)
#         outputs = model(Xb)

#         outputs = torch.sigmoid(outputs.squeeze())

#         loss = criterion(outputs, yb)
#         loss.backward()
#         optimizer.step()

#         total_loss += loss.item()

#         # ✅ Accuracy calculation (no need for torch.no_grad here)
#         predicted = (outputs > 0.5).float()

#         tot_vals += yb.size(0)
#         correct_vals += (predicted == yb).sum().item()

#     acc = correct_vals / tot_vals * 100
#     avg_loss = total_loss / len(train_loader)

#     print(f"Epoch {epoch+1}/{epochs}, Loss = {avg_loss:.4f}, Accuracy = {acc:.2f}%")
```

```python
epochs = 15

for epoch in range(epochs):
    model.train()

    correct_vals = 0
    tot_vals = 0
    total_loss = 0

    # 👉 counters for precision & recall
    TP, FP, FN = 0, 0, 0

    for Xb, yb in train_loader:
        optimizer.zero_grad()

        Xb = Xb.unsqueeze(1)
        outputs = model(Xb)

        outputs = torch.sigmoid(outputs.squeeze())

        loss = criterion(outputs, yb)
        loss.backward()
        optimizer.step()

        total_loss += loss.item()

        predicted = (outputs > 0.6).float()

        # ✅ Accuracy
        tot_vals += yb.size(0)
        correct_vals += (predicted == yb).sum().item()

        # ✅ Precision & Recall components
        TP += ((predicted == 1) & (yb == 1)).sum().item()
        FP += ((predicted == 1) & (yb == 0)).sum().item()
        FN += ((predicted == 0) & (yb == 1)).sum().item()

    # ✅ Metrics
    acc = correct_vals / tot_vals * 100
    avg_loss = total_loss / len(train_loader)

    precision = TP / (TP + FP) if (TP + FP) > 0 else 0
    recall = TP / (TP + FN) if (TP + FN) > 0 else 0

    print(f"Epoch {epoch+1}/{epochs}")
    print(f"Loss = {avg_loss:.4f}, Accuracy = {acc:.2f}%")
    print(f"Precision = {precision:.4f}, Recall = {recall:.4f}\n")
```

```python
# evaluate

model.eval()

with torch.no_grad():
    correct_vals = 0
    tot_vals = 0

    # 👉 new counters
    TP, FP, FN = 0, 0, 0

    for Xb, yb in test_loader:
        Xb = Xb.unsqueeze(1)

        outputs = model(Xb)
        predicted = (torch.sigmoid(outputs.squeeze()) > 0.6).float()

        # ✅ Accuracy
        tot_vals += yb.size(0)
        correct_vals += (predicted == yb).sum().item()

        # ✅ Precision & Recall components
        TP += ((predicted == 1) & (yb == 1)).sum().item()
        FP += ((predicted == 1) & (yb == 0)).sum().item()
        FN += ((predicted == 0) & (yb == 1)).sum().item()

    # ✅ Final metrics
    accuracy = correct_vals / tot_vals * 100
    precision = TP / (TP + FP) if (TP + FP) > 0 else 0
    recall = TP / (TP + FN) if (TP + FN) > 0 else 0

    print(f"Accuracy = {accuracy:.2f}%")
    print(f"Precision = {precision:.4f}")
    print(f"Recall = {recall:.4f}")
```

```python
def predict_sentiment(text):
    text = text.lower()

    vectorized = tf.transform([text]).toarray()
    X = torch.tensor(vectorized, dtype=torch.float32).unsqueeze(1)

    with torch.no_grad():
        output = model(X)
        prob = torch.sigmoid(output.squeeze()).item()

    sentiment = "Positive 😊" if prob > 0.6 else "Negative 😞"
    return sentiment, prob
```

```python
sentences = [
    "I love this product, it is amazing!",
    "This is the worst experience ever",
    "The movie was okay, not great",
    "Absolutely fantastic service!",
    "I hate this, very disappointing"
]

for s in sentences:
    result, prob = predict_sentiment(s)
    print(f"Sentence: {s}")
    print(f"Prediction: {result} (Confidence: {prob:.4f})\n")
```
